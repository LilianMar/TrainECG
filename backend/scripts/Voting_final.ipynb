```python
# voting_with_resnet_finetune_corrected.py
# Script corregido manteniendo la funcionalidad original exitosa
# Solo implementando mejoras específicas solicitadas
# Requisitos: tensorflow, sklearn, opencv-python, matplotlib, seaborn, pandas

import os
import random
import time
import warnings
from collections import Counter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_curve, auc, confusion_matrix, f1_score, accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_class_weight

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical

import cv2

warnings.filterwarnings('ignore')

# -------------------------
# Configuración global (CONSERVADA DEL ORIGINAL)
# -------------------------
SEED = 42
IMG_SIZE = (128, 128)
BATCH_SIZE = 32
EPOCHS = 50
MAX_PER_CLASS = 2000

TRAIN_SPLIT = 0.70
VAL_SPLIT = 0.15
TEST_SPLIT = 0.15

np.random.seed(SEED)
tf.random.set_seed(SEED)
random.seed(SEED)

print("Script: voting_with_resnet_finetune_corrected.py")
print(f"IMG_SIZE={IMG_SIZE}, MAX_PER_CLASS={MAX_PER_CLASS}, EPOCHS={EPOCHS}")

# ===========================
# Focal Loss (ORIGINAL - SIN CAMBIOS)
# ===========================
def focal_loss(gamma=2., alpha=None):
    """Focal loss factory. Devuelve una función de pérdida que Keras puede usar."""
    def focal(y_true, y_pred):
        eps = 1e-7
        y_pred = tf.clip_by_value(y_pred, eps, 1. - eps)
        ce = - y_true * tf.math.log(y_pred)
        if alpha is not None:
            alpha_tensor = tf.constant(alpha, dtype=tf.float32)
            alpha_weight = tf.reduce_sum(alpha_tensor * y_true, axis=1)
            weight = alpha_weight[:, None] * tf.pow(1 - y_pred, gamma)
        else:
            weight = tf.pow(1 - y_pred, gamma)
        fl = weight * ce
        return tf.reduce_sum(fl, axis=1)
    return focal

# ===========================
# Data processor (CONSERVADO DEL ORIGINAL + VISUALIZACIÓN AGREGADA)
# ===========================
class ECGDataProcessor:
    def __init__(self, data_path, img_size=(128,128), max_per_class=2000):
        self.data_path = data_path
        self.img_size = img_size
        self.max_per_class = max_per_class
        self.label_encoder = LabelEncoder()
        self.class_distribution = {}
        self.sample_images = {}  # NUEVO: para visualización

    def load_and_limit_by_max(self):
        class_images = {}
        print("\nCargando imágenes (limitando por clase):")
        for name in sorted(os.listdir(self.data_path)):
            full = os.path.join(self.data_path, name)
            if not os.path.isdir(full):
                continue
            files = [f for f in os.listdir(full) if f.lower().endswith(('.png','.jpg','.jpeg'))]
            files_paths = [os.path.join(full, f) for f in files]
            if len(files_paths) > self.max_per_class:
                files_paths = random.sample(files_paths, self.max_per_class)
                print(f"  {name}: muestreo a {self.max_per_class}")
            else:
                print(f"  {name}: {len(files_paths)}")
            class_images[name] = files_paths
            self.class_distribution[name] = len(files_paths)
            
            # NUEVO: Guardar 3 muestras para visualización
            if len(files_paths) >= 3:
                self.sample_images[name] = random.sample(files_paths, 3)
            else:
                self.sample_images[name] = files_paths
                
        total = sum(len(v) for v in class_images.values())
        print("Total imágenes consideradas:", total)
        return class_images

    def preprocess_image(self, img_path):
        """CONSERVADO DEL ORIGINAL - SIN CAMBIOS"""
        try:
            img = cv2.imread(img_path)
            if img is None:
                return None
            if len(img.shape) == 3:
                img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            img = cv2.resize(img, self.img_size, interpolation=cv2.INTER_AREA)
            img = cv2.GaussianBlur(img, (3,3), 0)
            img = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)
            _, thr = cv2.threshold(img.astype(np.uint8), 0, 255, cv2.THRESH_OTSU)
            final = thr.astype(np.float32) / 255.0
            final = np.expand_dims(final, axis=-1)
            return final
        except Exception as e:
            print("Error preprocess:", img_path, e)
            return None

    def show_sample_images(self):
        """NUEVO: Mostrar 3 imágenes aleatorias por clase"""
        print("\nMOSTRANDO MUESTRAS PREPROCESADAS POR CLASE")
        print("=" * 60)
        
        num_classes = len(self.sample_images)
        fig, axes = plt.subplots(num_classes, 3, figsize=(12, 3 * num_classes))
        fig.suptitle('Muestras Preprocesadas por Clase ECG', fontsize=16, fontweight='bold')
        
        if num_classes == 1:
            axes = axes.reshape(1, -1)
        
        for class_idx, (class_name, img_paths) in enumerate(self.sample_images.items()):
            for img_idx, img_path in enumerate(img_paths):
                processed_img = self.preprocess_image(img_path)
                if processed_img is not None:
                    axes[class_idx, img_idx].imshow(processed_img.squeeze(), cmap='gray')
                    axes[class_idx, img_idx].set_title(f'Clase {class_name}')
                    axes[class_idx, img_idx].axis('off')
                else:
                    axes[class_idx, img_idx].text(0.5, 0.5, 'Error\ncargando', 
                                                ha='center', va='center')
                    axes[class_idx, img_idx].axis('off')
        
        plt.tight_layout()
        plt.show()
        
        # Mostrar distribución de clases
        plt.figure(figsize=(10, 6))
        classes = list(self.class_distribution.keys())
        counts = list(self.class_distribution.values())
        
        # Gráfico horizontal como solicitado
        plt.barh(classes, counts)
        plt.title('Distribución de Clases en el Dataset ECG', fontsize=14, fontweight='bold')
        plt.xlabel('Número de Imágenes')
        plt.ylabel('Clases ECG')
        
        # Agregar valores en las barras
        for i, count in enumerate(counts):
            plt.text(count + 10, i, str(count), ha='left', va='center', fontweight='bold')
        
        plt.tight_layout()
        plt.show()

    def prepare_dataset(self):
        class_images = self.load_and_limit_by_max()
        
        # NUEVO: Mostrar muestras
        self.show_sample_images()
        
        X, y = [], []
        for cls, paths in class_images.items():
            for p in paths:
                proc = self.preprocess_image(p)
                if proc is not None:
                    X.append(proc)
                    y.append(cls)
        X = np.array(X, dtype=np.float32)
        y = np.array(y)
        y_enc = self.label_encoder.fit_transform(y)
        y_onehot = to_categorical(y_enc)
        class_names = list(self.label_encoder.classes_)
        print(f"\nDataset: X={X.shape}, y={y_onehot.shape}, clases={class_names}")
        return X, y_onehot, class_names

# ===========================
# Augmentación dirigida (CONSERVADO DEL ORIGINAL)
# ===========================
def augment_class_examples(X, y_onehot, class_names, target_label='N', augment_factor=2):
    if target_label not in class_names:
        return X, y_onehot
    idx_target = class_names.index(target_label)
    y_labels = np.argmax(y_onehot, axis=1)
    idxs = np.where(y_labels == idx_target)[0]
    if len(idxs) == 0:
        return X, y_onehot

    datagen = ImageDataGenerator(
        rotation_range=8,
        width_shift_range=0.05,
        height_shift_range=0.05,
        zoom_range=0.08,
        shear_range=0.04,
        brightness_range=(0.9,1.1),
        fill_mode='nearest'
    )
    X_new, y_new = [], []
    for i in idxs:
        img = X[i]
        batch = np.expand_dims(img, 0)
        it = datagen.flow(batch, batch_size=1)
        for _ in range(augment_factor):
            aug = next(it)[0]
            X_new.append(aug)
            y_new.append(y_onehot[i])
    if len(X_new) > 0:
        X = np.concatenate([X, np.array(X_new)], axis=0)
        y_onehot = np.concatenate([y_onehot, np.array(y_new)], axis=0)
        print(f"Augmented {len(X_new)} samples for class {target_label}")
    return X, y_onehot

# ===========================
# Capas y métricas (CONSERVADAS DEL ORIGINAL)
# ===========================
class AttentionLayer(layers.Layer):
    def __init__(self, units=128, **kwargs):
        super().__init__(**kwargs)
        self.units = units
    def build(self, input_shape):
        self.W_q = self.add_weight(shape=(input_shape[-1], self.units), initializer='glorot_uniform')
        self.W_k = self.add_weight(shape=(input_shape[-1], self.units), initializer='glorot_uniform')
        self.W_v = self.add_weight(shape=(input_shape[-1], self.units), initializer='glorot_uniform')
        self.dense = layers.Dense(input_shape[-1])
        super().build(input_shape)
    def call(self, inputs):
        Q = tf.matmul(inputs, self.W_q)
        K = tf.matmul(inputs, self.W_k)
        V = tf.matmul(inputs, self.W_v)
        attn = tf.matmul(Q, K, transpose_b=True)
        attn = attn / tf.sqrt(tf.cast(self.units, tf.float32))
        w = tf.nn.softmax(attn, axis=-1)
        out = tf.matmul(w, V)
        return self.dense(out) + inputs

class SpatialAttentionLayer(layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
    def build(self, input_shape):
        self.conv1 = layers.Conv2D(64, (3,3), padding='same', activation='relu')
        self.conv2 = layers.Conv2D(1, (1,1), padding='same', activation='sigmoid')
        super().build(input_shape)
    def call(self, x):
        m = self.conv1(x)
        m = self.conv2(m)
        return x * m

class F1Score(tf.keras.metrics.Metric):
    def __init__(self, name='f1_score', **kwargs):
        super().__init__(name=name, **kwargs)
        self.prec = tf.keras.metrics.Precision()
        self.rec = tf.keras.metrics.Recall()
    def update_state(self, y_true, y_pred, sample_weight=None):
        self.prec.update_state(y_true, y_pred, sample_weight)
        self.rec.update_state(y_true, y_pred, sample_weight)
    def result(self):
        p = self.prec.result()
        r = self.rec.result()
        return 2 * ((p * r) / (p + r + tf.keras.backend.epsilon()))
    def reset_state(self):
        self.prec.reset_state()
        self.rec.reset_state()

# ===========================
# Modelos corregidos
# ===========================

class SimpleECGModel:
    """CONSERVADO DEL ORIGINAL - SIN CAMBIOS"""
    def __init__(self, input_shape, num_classes): 
        self.input_shape=input_shape; self.num_classes=num_classes
    def build_model(self):
        inp = layers.Input(shape=self.input_shape)
        x = layers.Conv2D(16,3,padding='same',activation='relu')(inp)
        x = layers.BatchNormalization()(x); x = layers.MaxPooling2D(2)(x)
        x = layers.Conv2D(32,3,padding='same',activation='relu')(x)
        x = layers.BatchNormalization()(x); x = layers.MaxPooling2D(2)(x)
        x = layers.Conv2D(64,3,padding='same',activation='relu')(x)
        x = layers.BatchNormalization()(x); x = layers.GlobalAveragePooling2D()(x)
        x = layers.Dropout(0.3)(x); x = layers.Dense(128,activation='relu')(x); x = layers.Dropout(0.2)(x)
        out = layers.Dense(self.num_classes, activation='softmax')(x)
        return models.Model(inp,out)

class ImprovedCNNModel:
    """NUEVO: CNN según especificaciones del usuario (corregida para 128x128)"""
    def __init__(self, input_shape, num_classes):
        self.input_shape = input_shape
        self.num_classes = num_classes

    def build_model(self):
        model = Sequential([
            # Etapa 1
            Conv2D(32, (3, 3), activation='relu', input_shape=self.input_shape),
            MaxPooling2D((2, 2)),  # 128x128 -> 64x64
            
            # Etapa 2  
            Conv2D(32, (3, 3), activation='relu'),
            MaxPooling2D((2, 2)),  # 64x64 -> 32x32
            
            # Etapa 3
            Conv2D(64, (3, 3), activation='relu'),
            MaxPooling2D((2, 2)),  # 32x32 -> 16x16
            
            # Etapa 4
            Conv2D(64, (3, 3), activation='relu'),
            MaxPooling2D((2, 2)),  # 16x16 -> 8x8
            
            # Etapa 5 - CORREGIDA: Sin MaxPooling para evitar que se vuelva demasiado pequeña
            Conv2D(128, (3, 3), activation='relu'),  # 8x8 -> 6x6
            
            # Etapa 6 - CORREGIDA: Sin MaxPooling para evitar dimensiones negativas
            Conv2D(128, (3, 3), activation='relu'),  # 6x6 -> 4x4
            
            # Global Average Pooling en lugar del último MaxPooling + Flatten
            layers.GlobalAveragePooling2D(),
            
            # Clasificador
            Dense(256, activation='relu'),
            Dropout(0.5),
            Dense(self.num_classes, activation='softmax')
        ])
        return model

class LSTMWithAttentionModel:
    """MODIFICADO: Peso especial para clase N en loss"""
    def __init__(self, input_shape, num_classes): 
        self.input_shape=input_shape; self.num_classes=num_classes
    def build_model(self):
        inputs = layers.Input(shape=self.input_shape)
        x = layers.Reshape((self.input_shape[0], self.input_shape[1]))(inputs)
        x = layers.LSTM(128, return_sequences=True, dropout=0.25, recurrent_dropout=0.15)(x)
        att = layers.Dense(1, activation='tanh')(x); att = layers.Flatten()(att); att = layers.Activation('softmax')(att)
        att = layers.RepeatVector(128)(att); att = layers.Permute([2,1])(att); x = layers.multiply([x, att])
        x = layers.LSTM(64, dropout=0.25, recurrent_dropout=0.15)(x); x = layers.Dense(128, activation='relu')(x)
        x = layers.Dropout(0.3)(x); out = layers.Dense(self.num_classes, activation='softmax')(x)
        return models.Model(inputs, out)

class HybridCNNLSTMWithAttentionModel:
    """MODIFICADO: Peso especial para clase N en loss"""
    def __init__(self, input_shape, num_classes): 
        self.input_shape=input_shape; self.num_classes=num_classes
    def build_model(self):
        inputs = layers.Input(shape=self.input_shape)
        x = layers.Conv2D(32,3,padding='same', activation='relu')(inputs); x = layers.BatchNormalization()(x); x = layers.MaxPooling2D(2)(x)
        x = SpatialAttentionLayer()(x); x = layers.Dropout(0.2)(x)
        x = layers.Conv2D(64,3,padding='same', activation='relu')(x); x = layers.BatchNormalization()(x); x = layers.MaxPooling2D(2)(x)
        x = SpatialAttentionLayer()(x); x = layers.Dropout(0.25)(x)
        x = layers.Conv2D(128,3,padding='same', activation='relu')(x); x = layers.BatchNormalization()(x); x = layers.MaxPooling2D(2)(x)
        x = layers.Dropout(0.2)(x)
        s = x.shape; x = layers.Reshape((s[1], s[2]*s[3]))(x)
        x = layers.Bidirectional(layers.LSTM(96, return_sequences=True, dropout=0.2, recurrent_dropout=0.15))(x)
        attended = AttentionLayer(units=192)(x); x = layers.GlobalAveragePooling1D()(attended)
        x = layers.Dense(256, activation='relu')(x); x = layers.Dropout(0.3)(x)
        x = layers.Dense(128, activation='relu')(x); x = layers.Dropout(0.25)(x)
        x = layers.Dense(64, activation='relu')(x); x = layers.Dropout(0.2)(x)
        out = layers.Dense(self.num_classes, activation='softmax')(x)
        return models.Model(inputs,out)

class ResNet50WithSpatialAttention:
    """CORREGIDO: Manejo de canal único para ResNet50"""
    def __init__(self, input_shape, num_classes, trainable_from=None):
        self.input_shape = input_shape
        self.num_classes = num_classes
        self.trainable_from = trainable_from
    def build_model(self):
        inputs = layers.Input(shape=self.input_shape)
        
        # Convertir de 1 canal a 3 canales para ResNet50
        x = layers.Conv2D(3, (1,1), padding='same')(inputs)
        
        base = tf.keras.applications.ResNet50(include_top=False, weights='imagenet', 
                                            input_shape=(self.input_shape[0], self.input_shape[1], 3))
        if self.trainable_from is None:
            for layer in base.layers:
                layer.trainable = False
        else:
            for i, layer in enumerate(base.layers):
                layer.trainable = True if i >= self.trainable_from else False

        x = base(x, training=False)
        x = SpatialAttentionLayer()(x)
        x = layers.GlobalAveragePooling2D()(x)
        x = layers.Dense(256, activation='relu')(x); x = layers.Dropout(0.4)(x)
        x = layers.Dense(128, activation='relu')(x); x = layers.Dropout(0.3)(x)
        outputs = layers.Dense(self.num_classes, activation='softmax')(x)
        return models.Model(inputs, outputs)

# ===========================
# Visualización (MEJORADA: Gráficos horizontales)
# ===========================
def plot_training_history(history, model_name="Modelo"):
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    best_epoch = np.argmin(history.history['val_loss']) + 1 if 'val_loss' in history.history else 1

    axes[0,0].plot(history.history['loss'], label='Training Loss')
    axes[0,0].plot(history.history['val_loss'], label='Validation Loss')
    axes[0,0].axvline(best_epoch-1, color='green', linestyle='--', label=f'Best Epoch ({best_epoch})')
    axes[0,0].set_title('Pérdida'); axes[0,0].legend(); axes[0,0].grid(True)

    axes[0,1].plot(history.history.get('accuracy', []), label='Training Acc')
    axes[0,1].plot(history.history.get('val_accuracy', []), label='Validation Acc')
    axes[0,1].axvline(best_epoch-1, color='green', linestyle='--'); axes[0,1].set_title('Exactitud'); axes[0,1].legend(); axes[0,1].grid(True)

    axes[1,0].plot(history.history.get('precision', []), label='Train Precision')
    axes[1,0].plot(history.history.get('val_precision', []), label='Val Precision'); axes[1,0].set_title('Precisión'); axes[1,0].legend(); axes[1,0].grid(True)

    axes[1,1].plot(history.history.get('recall', []), label='Train Recall')
    axes[1,1].plot(history.history.get('val_recall', []), label='Val Recall'); axes[1,1].set_title('Recall'); axes[1,1].legend(); axes[1,1].grid(True)

    plt.suptitle(f'{model_name} - Historial de Entrenamiento'); plt.tight_layout(); plt.show()

def plot_confusion_matrix_normalized(y_true, y_pred, class_names, title="Matriz de Confusión (normalizada)"):
    cm = confusion_matrix(y_true, y_pred)
    cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.figure(figsize=(9,7))
    sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title(title); plt.xlabel('Predicción'); plt.ylabel('Etiqueta real'); plt.show()

def plot_comparison_horizontal(results_df):
    """NUEVO: Gráficos de comparación horizontales"""
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    models = results_df['Modelo']
    y_pos = np.arange(len(models))
    
    # F1-Score Macro (horizontal)
    bars1 = axes[0,0].barh(y_pos, results_df['F1_Macro'])
    axes[0,0].set_yticks(y_pos)
    axes[0,0].set_yticklabels(models)
    axes[0,0].set_xlabel('F1-Score Macro')
    axes[0,0].set_title('F1-Score Macro por Modelo')
    axes[0,0].grid(axis='x', alpha=0.3)
    
    for i, v in enumerate(results_df['F1_Macro']):
        axes[0,0].text(v + 0.01, i, f'{v:.3f}', va='center')

    # Accuracy (horizontal)
    bars2 = axes[0,1].barh(y_pos, results_df['Accuracy'])
    axes[0,1].set_yticks(y_pos)
    axes[0,1].set_yticklabels(models)
    axes[0,1].set_xlabel('Accuracy')
    axes[0,1].set_title('Accuracy por Modelo')
    axes[0,1].grid(axis='x', alpha=0.3)
    
    for i, v in enumerate(results_df['Accuracy']):
        axes[0,1].text(v + 0.01, i, f'{v:.3f}', va='center')

    # Tiempo (horizontal)
    bars3 = axes[1,0].barh(y_pos, results_df['Tiempo_min'])
    axes[1,0].set_yticks(y_pos)
    axes[1,0].set_yticklabels(models)
    axes[1,0].set_xlabel('Tiempo de Entrenamiento (min)')
    axes[1,0].set_title('Tiempo de Entrenamiento')
    axes[1,0].grid(axis='x', alpha=0.3)
    
    for i, v in enumerate(results_df['Tiempo_min']):
        axes[1,0].text(v + 1, i, f'{v:.1f}', va='center')

    # Parámetros (horizontal, escala log)
    bars4 = axes[1,1].barh(y_pos, results_df['Params'])
    axes[1,1].set_yticks(y_pos)
    axes[1,1].set_yticklabels(models)
    axes[1,1].set_xlabel('Número de Parámetros')
    axes[1,1].set_title('Complejidad del Modelo')
    axes[1,1].set_xscale('log')
    axes[1,1].grid(axis='x', alpha=0.3)
    
    for i, v in enumerate(results_df['Params']):
        axes[1,1].text(v * 1.1, i, f'{v:,}', va='center', fontsize=9)

    plt.suptitle('Comparación de Modelos - Gráficos Horizontales')
    plt.tight_layout()
    plt.show()

# ===========================
# Entrenamiento (MODIFICADO: Solo peso balanceado + alpha moderada para N)
# ===========================
def train_and_evaluate_model(model_class, model_name, X_train, X_val, X_test, y_train, y_val, y_test,
                             class_names, learning_rate=3e-4, epochs=EPOCHS, save_model=True,
                             augment_N=True, augment_factor=2, model_kwargs=None, batch_size=BATCH_SIZE):
    print(f"\n=== Entrenando: {model_name} ===")
    if augment_N:
        X_train_aug, y_train_aug = augment_class_examples(X_train, y_train, class_names, target_label='N', augment_factor=augment_factor)
    else:
        X_train_aug, y_train_aug = X_train, y_train

    y_train_labels = np.argmax(y_train_aug, axis=1)
    classes = np.unique(y_train_labels)
    weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train_labels)
    class_weight_dict = {int(i): float(w) for i, w in enumerate(weights)}
    
    # MODIFICADO: Boost moderado para N en TODOS los modelos
    if 'N' in class_names:
        n_idx = class_names.index('N')
        class_weight_dict[n_idx] *= 2  # Boost moderado para clase N en todos los modelos
        print(f"Boost moderado para clase N aplicado: {class_weight_dict[n_idx]:.3f}")
    
    print("Class weights:", class_weight_dict)

    kw = model_kwargs or {}
    model_builder = model_class(X_train_aug.shape[1:], len(class_names), **kw)
    model = model_builder.build_model()

    # MODIFICADO: Alpha moderada para N en focal loss para TODOS los modelos
    alpha = [1.0]*len(class_names)
    if 'N' in class_names:
        alpha[class_names.index('N')] = 2  # Boost moderado para todos los modelos

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
        loss=focal_loss(gamma=2.0, alpha=alpha),
        metrics=['accuracy', keras.metrics.Precision(name='precision'), keras.metrics.Recall(name='recall'), F1Score()]
    )

    print(f"Parámetros totales: {model.count_params():,}")

    model_filename = f"best_model_{model_name}.h5"
    cb = [
        callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1),
        callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-7, verbose=1)
    ]
    if save_model:
        mc = callbacks.ModelCheckpoint(filepath=model_filename, monitor='val_f1_score', mode='max',
                                       save_best_only=True, save_weights_only=False, verbose=1)
        cb.append(mc)

    start = time.time()
    history = model.fit(X_train_aug, y_train_aug,
                        validation_data=(X_val, y_val),
                        epochs=epochs,
                        batch_size=batch_size,
                        class_weight=class_weight_dict,
                        callbacks=cb,
                        verbose=1)
    elapsed = time.time() - start

    # CORREGIDO: Carga del modelo guardado
    if save_model and os.path.exists(model_filename):
        try:
            # Objetos personalizados corregidos - función focal_loss debe ser callable
            def focal_loss_custom(gamma=2., alpha=alpha):
                def focal(y_true, y_pred):
                    eps = 1e-7
                    y_pred = tf.clip_by_value(y_pred, eps, 1. - eps)
                    ce = - y_true * tf.math.log(y_pred)
                    if alpha is not None:
                        alpha_tensor = tf.constant(alpha, dtype=tf.float32)
                        alpha_weight = tf.reduce_sum(alpha_tensor * y_true, axis=1)
                        weight = alpha_weight[:, None] * tf.pow(1 - y_pred, gamma)
                    else:
                        weight = tf.pow(1 - y_pred, gamma)
                    fl = weight * ce
                    return tf.reduce_sum(fl, axis=1)
                return focal
            
            custom_objects = {
                'AttentionLayer': AttentionLayer,
                'SpatialAttentionLayer': SpatialAttentionLayer,
                'F1Score': F1Score,
                'focal_loss': focal_loss_custom(),
                'focal': focal_loss_custom()
            }
            model = keras.models.load_model(model_filename, custom_objects=custom_objects)
            print("Cargado modelo guardado:", model_filename)
        except Exception as e:
            print("No se cargó modelo guardado:", e)

    # NUEVO: Evaluación INMEDIATA en test con classification report
    print(f"\n{'='*60}")
    print(f"EVALUACIÓN EN CONJUNTO DE PRUEBA - {model_name}")
    print(f"{'='*60}")
    
    y_pred_proba = model.predict(X_test, verbose=0)
    y_pred = np.argmax(y_pred_proba, axis=1)
    y_true = np.argmax(y_test, axis=1)
    
    # Métricas finales
    acc = accuracy_score(y_true, y_pred)
    f1_macro = f1_score(y_true, y_pred, average='macro')
    f1_weighted = f1_score(y_true, y_pred, average='weighted')
    
    print(f"MÉTRICAS EN TEST:")
    print(f"  • Accuracy: {acc:.4f}")
    print(f"  • F1-Macro: {f1_macro:.4f}")
    print(f"  • F1-Weighted: {f1_weighted:.4f}")
    
    print(f"\nCLASSIFICATION REPORT - {model_name} (CONJUNTO DE PRUEBA):")
    print("=" * 70)
    report = classification_report(y_true, y_pred, target_names=class_names, digits=4)
    print(report)
    print("=" * 70)

    # Visualizaciones
    print(f"Generando visualizaciones para {model_name}...")
    plot_training_history(history, model_name)
    plot_confusion_matrix_normalized(y_true, y_pred, class_names, title=f"Matriz Confusión - {model_name}")

    return {
        'model': model,
        'history': history,
        'accuracy': acc,
        'f1_macro': f1_macro,
        'f1_weighted': f1_weighted,
        'training_time': elapsed,
        'epochs_trained': len(history.history['loss']),
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba,
        'y_true': y_true,
        'model_filename': model_filename if save_model else None
    }

# ===========================
# Data prep (CONSERVADO)
# ===========================
def prepare_data_unified(data_path):
    if not os.path.exists(data_path):
        print("Ruta no encontrada:", data_path)
        return (None,)*7
    proc = ECGDataProcessor(data_path, IMG_SIZE, max_per_class=MAX_PER_CLASS)
    X, y, class_names = proc.prepare_dataset()
    X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=TEST_SPLIT, random_state=SEED, stratify=y)
    val_adj = VAL_SPLIT / (1 - TEST_SPLIT)
    X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=val_adj, random_state=SEED, stratify=y_temp)
    print(f"Split sizes -> train: {X_train.shape[0]}, val: {X_val.shape[0]}, test: {X_test.shape[0]}")
    return X_train, X_val, X_test, y_train, y_val, y_test, class_names

# ===========================
# Pipeline principal (CORREGIDO)
# ===========================
def run_all_models_comparison_corrected():
    DATA_PATH = "ecg_image_data"
    data = prepare_data_unified(DATA_PATH)
    if data[0] is None:
        return
    X_train, X_val, X_test, y_train, y_val, y_test, class_names = data

    models_to_test = [
        (SimpleECGModel, "Simple_CNN", 3e-4, {}),
        (ImprovedCNNModel, "CNN_Mejorada_Usuario", 3e-4, {}),  # NUEVO: Tu CNN específica
        (LSTMWithAttentionModel, "LSTM_Attention", 3e-4, {}),
        (HybridCNNLSTMWithAttentionModel, "Hybrid_CNN_LSTM_Attention", 3e-4, {}),
    ]

    results = {}

    # Entrenar modelos no-ResNet
    for ModelClass, name, lr, kw in models_to_test:
        try:
            res = train_and_evaluate_model(ModelClass, name,
                                           X_train, X_val, X_test, y_train, y_val, y_test,
                                           class_names, learning_rate=lr, epochs=EPOCHS,
                                           save_model=True, augment_N=True, augment_factor=2, model_kwargs=kw)
            results[name] = res
        except tf.errors.ResourceExhaustedError as oom:
            print(f"OOM durante entrenamiento de {name}: {oom}. Reducir MAX_PER_CLASS o batch_size.")
        except Exception as e:
            print(f"Error entrenando {name}: {e}")

    # ResNet50 con batch size menor
    try:
        print("\n=== Entrenando ResNet50 (batch size reducido) ===")
        res_resnet = train_and_evaluate_model(ResNet50WithSpatialAttention, "ResNet50_Spatial",
                                            X_train, X_val, X_test, y_train, y_val, y_test,
                                            class_names, learning_rate=1e-4, epochs=25,
                                            save_model=True, augment_N=True, augment_factor=2,
                                            model_kwargs={'trainable_from': 140}, batch_size=16)
        results["ResNet50_Spatial"] = res_resnet
    except tf.errors.ResourceExhaustedError as oom:
        print("OOM durante ResNet50:", oom)
    except Exception as e:
        print("Error ResNet50:", e)

    # Compilar resultados finales
    print(f"\n{'='*80}")
    print("COMPILANDO RESULTADOS FINALES DE TODOS LOS MODELOS")
    print(f"{'='*80}")
    
    rows = []
    for k, v in results.items():
        if v is None:
            print(f"Saltando {k} - no hay resultados")
            continue
            
        model_obj = v.get('model')
        rows.append({
            'Modelo': k,
            'Accuracy': v.get('accuracy', 0.0),
            'F1_Macro': v.get('f1_macro', 0.0),
            'F1_Weighted': v.get('f1_weighted', 0.0),
            'Tiempo_min': v.get('training_time', 0)/60,
            'Epochs': v.get('epochs_trained', 0),
            'Params': model_obj.count_params() if model_obj is not None else 0,
            'SavedModel': v.get('model_filename', 'N/A')
        })
    
    if len(rows) == 0:
        print("ERROR: No hay resultados para mostrar")
        return None, None
    
    df = pd.DataFrame(rows).sort_values('F1_Macro', ascending=False)
    
    print(f"\n{'='*80}")
    print("TABLA COMPARATIVA FINAL - ORDENADA POR F1-MACRO")
    print(f"{'='*80}")
    print(df.to_string(index=False))
    
    # Guardar CSV
    csv_filename = "model_comparison_corrected.csv"
    df.to_csv(csv_filename, index=False)
    print(f"\nResultados guardados en: {csv_filename}")
    
    # Mostrar mejor modelo
    best_model = df.iloc[0]
    print(f"\n🏆 MEJOR MODELO: {best_model['Modelo']}")
    print(f"   • F1-Macro: {best_model['F1_Macro']:.4f}")
    print(f"   • Accuracy: {best_model['Accuracy']:.4f}")
    print(f"   • Tiempo: {best_model['Tiempo_min']:.1f} min")
    
    # Generar gráficos de comparación
    print(f"\nGenerando gráficos comparativos...")
    plot_comparison_horizontal(df)
    
    print(f"{'='*80}")
    print("ANÁLISIS COMPLETADO EXITOSAMENTE")
    print(f"{'='*80}")
    
    return results, df

# ===========================
# MAIN
# ===========================
if __name__ == "__main__":
    # Configurar TensorFlow para M2
    try:
        gpus = tf.config.experimental.list_physical_devices('GPU')
        if gpus:
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, True)
    except Exception as e:
        print(f"Advertencia TF config: {e}")
    
    DATA_PATH = "ecg_image_data"
    if not os.path.exists(DATA_PATH):
        print("ERROR: no existe la ruta de datos:", DATA_PATH)
        exit(1)

    print("Ejecutando versión corregida del script...")
    results, df = run_all_models_comparison_corrected()
```
